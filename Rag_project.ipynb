```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    separators=["\n\n", "\n", " "]
)
```

```python
chunks = recursive_splitter.split_documents(loaded_pdf_doc)
print(len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import Chroma
vector_store = Chroma.from_documents(chunks,embeddings)
```

```python
vector_index=vector_store.as_retriever(search_kwargs={'k':5})
```

```python
pip install langchain-classic
```

```python
from langchain_classic.chains import RetrievalQA
rag = RetrievalQA.from_chain_type(
    model,
    retriever = vector_index,
    return_source_documents=True
)
```

```python
question = 'Am I eligible to apply for the MakeMyTrip India’s Favourite Homestay Awards, and which award categories does my property qualify for based on the eligibility criteria?'
response = rag({"query":question})
print(response)
```

```python
question = "Am I eligible to apply for the MakeMyTrip India’s Favourite Homestay Awards, and which award categories does my property qualify for based on the eligibility criteria?"

response = rag.invoke({"query": question})

print(response["result"])
```
